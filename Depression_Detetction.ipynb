```python
import os
import mne
import pandas as pd

data_folder = '/content/drive/My Drive/4244171'
#List of all EDF file
edf_files = [f for f in os.listdir(data_folder) if f.endswith('.edf')]

all_data = pd.DataFrame()

for file in edf_files:
    file_path = os.path.join(data_folder, file)  #Combine the data_folder path with current path
    raw = mne.io.read_raw_edf(file_path, preload=True)

    data, times = raw.get_data(), raw.times

    df = pd.DataFrame(data.T, columns=raw.ch_names)
    df['Time'] = times

    # Append to the final DataFrame
    all_data = pd.concat([all_data, df], ignore_index=True)


```

```python
data_folder = '/content/drive/My Drive/4244171'

edf_files = [f for f in os.listdir(data_folder) if f.endswith('.edf')]

sampling_rates = []

for file in edf_files:
    file_path = os.path.join(data_folder, file)
    raw = mne.io.read_raw_edf(file_path, preload=True)

    # Get the sampling rate (sfreq) for the current file
    sfreq = raw.info['sfreq']
    sampling_rates.append(sfreq)

# Check if all sampling rates are consistent
unique_sampling_rates = set(sampling_rates)

if len(unique_sampling_rates) == 1:
    print(f"All files have the same sampling rate: {unique_sampling_rates.pop()}")
else:
    print(f"Sampling rates are inconsistent: {unique_sampling_rates}")
```
